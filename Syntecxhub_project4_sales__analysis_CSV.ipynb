 STEP 2 : Create Sales Data File (CSV File Creation)

```python
import pandas as pd

# Creating sample sales dataset
data = {
    'Order_ID': [101,102,103,104,105,106,107,108],
    'Product': ['Laptop','Mobile','Tablet','Laptop','Mobile','Tablet','Laptop','Mobile'],
    'Category': ['Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics','Electronics'],
    'Region': ['South','North','East','West','South','North','East','West'],
    'Sales': [50000,20000,15000,55000,22000,18000,52000,25000],
    'Quantity': [5,10,7,6,11,8,5,12],
    'Order_Date': pd.date_range(start='2024-01-01', periods=8)
}

df = pd.DataFrame(data)

# Save as CSV file
df.to_csv("sales_data.csv", index=False)

print("File Created Successfully ✅")
```

 This creates a file named sales_data.csv

 STEP 7: Sales by Product

```python
product_sales = df.groupby('Product')['Sales'].sum()
```

groupby('Product') → Groups same products.

.sum() → Adds sales for each product.

 Purpose: Compare product performance.


STEP 8: Sales by Region




```python
region_sales = df.groupby('Region')['Sales'].sum()
print(region_sales)
```

Groups data by region.

Calculates regional revenue.

 Purpose: Identify high-performing regions.



 STEP 9: Best Selling Product

```python
best_product = df.groupby('Product')['Quantity'].sum().idxmax()
print("Best Selling Product:", best_product)
```

STEP 10: Visualization



```python
import matplotlib.pyplot as plt

product_sales.plot(kind='bar')
plt.title("Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

 STEP 11: Monthly Sales Trend

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df['Month'] = df['Order_Date'].dt.month

monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
